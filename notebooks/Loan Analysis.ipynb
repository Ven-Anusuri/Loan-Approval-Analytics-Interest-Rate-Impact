```python
loan_clean = loan_df.copy()

# Fill nulls
median_income = loan_clean['Income'].median()
loan_clean['Income'] = loan_clean['Income'].fillna(median_income)

median_cs = loan_clean['CreditScore'].median()
loan_clean['CreditScore'] = loan_clean['CreditScore'].fillna(median_cs)

loan_clean['Education'] = loan_clean['Education'].fillna('Unknown')

# Remove invalid values — KEEP THE VALID ONES
loan_clean = loan_clean[loan_clean['Income'] > 0]
loan_clean = loan_clean[loan_clean['LoanAmount'] > 0]
loan_clean = loan_clean[(loan_clean['CreditScore'] >= 300) & (loan_clean['CreditScore'] <= 850)]

# Convert credit score to integer
loan_clean['CreditScore'] = loan_clean['CreditScore'].astype(int)

print(f"Final shape: {loan_clean.shape}")
print(f"Nulls remaining: {loan_clean.isnull().sum()}")
```

```python
start_date = pd.Timestamp('2018-01-01')
end_date= pd.Timestamp('2026-04-30')
total_days=(end_date-start_date).days
random_days=np.random.randint(0,total_days,size=len(loan_clean))
loan_clean['LoanDate']=start_date+ pd.to_timedelta(random_days,unit='D')
```

```python
correlation = loan_clean['CreditScore'].corr(loan_clean['LoanApproved'])
print(f"Correlation between CreditScore and LoanApproved: {correlation:.4f}")
```

```python
avg_credit_score_by_approval = loan_clean.groupby('LoanApproved')['CreditScore'].mean().reset_index()
print("Average CreditScore by Loan Approval Status:")
print(avg_credit_score_by_approval)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Define features (X) and target (y)
X = loan_clean[['CreditScore']]
y = loan_clean['LoanApproved']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(f"Training set size: {X_train.shape[0]} samples")
print(f"Test set size: {X_test.shape[0]} samples")
```

```python
# Initialize and train the logistic regression model
model = LogisticRegression(random_state=42)
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print(f"Model Accuracy: {accuracy:.4f}")
print("\nClassification Report:")
print(report)
print("\nConfusion Matrix:")
print(conf_matrix)

# Display model coefficients
print(f"\nCoefficient for CreditScore: {model.coef_[0][0]:.4f}")
print(f"Intercept: {model.intercept_[0]:.4f}")
```
